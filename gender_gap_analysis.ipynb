```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

Let's read the data into a dataframe and drop rows as matplotlib is unable to handle rows with missing values

```python
recent_grads=pd.read_csv('recent-grads.csv')
recent_grads.describe()
raw_data_count=recent_grads.shape[0]
raw_data_count
recent_grads=recent_grads.dropna()
cleaned_data_count=recent_grads.shape[0]
cleaned_data_count
```

Experimenting with wrappers in pandas for matplotlib 

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
ax.set_title('Employed vs. Sample_size')
```

```python
plot1=recent_grads.plot(x='Sample_size',y='Median',kind='scatter',title='Medain vs Sample Size')
```

```python
plot2 = recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Sample size vs Unemployment Rate')
```

```python
plot3=recent_grads.plot(x='Full_time',y='Median',kind='scatter',title='Full_time v/s Median')
```

```python
plot4=recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter',title='Sharewomen v/s Unemployment rate')
```

```python
recent_grads['Sample_size'].hist(bins=200,figsize=(10,10))
```

```python
recent_grads['Men'].hist(bins=100)
```

```python
cols=['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig=plt.figure(figsize=(20,40))
for r in range(1,4):
    ax = fig.add_subplot(8,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist',rot=40)
    ax.set_xlabel(cols[r])
    
```
